```python
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv"
dataset = pd.read_csv(url)
dataset
```

```python
#checking the shape of our data
dataset.shape
```

```python
#checking for null data
dataset.isnull().sum()
```

```python
"""fitting a linear model on the relationship between the temperature in the living room
and the temperature outside the building and getting the R^2 value"""
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lin = dataset[['T2','T6']]
linear_mod = LinearRegression()
X = lin[['T2']]
y = lin[['T6']]
model_1 = linear_mod.fit(X,y)
pred_1 = model_1.predict(X)
rsquare = round(r2_score(y, pred_1),2)
rsquare
```

```python
#dropping selected columns
dataset= dataset.drop(['date', 'lights'],axis = 1)
dataset
```

```python
#normalising our dataset using the min max scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(dataset),columns = dataset.columns)
```

```python
#splitting our dataset
from sklearn.model_selection import train_test_split
train = normalised_df.drop(['Appliances'], axis =1)
y = normalised_df['Appliances']
x_train, x_test, y_train, y_test = train_test_split(train, y, test_size = 0.3, random_state = 42)
```

```python
#training multiple linear regression
model_2 = LinearRegression()
model_2.fit(x_train, y_train)
pred_2 = model_2.predict(x_test)
```

```python
#calculating the mean absolute error in 2d.p
from sklearn.metrics import mean_absolute_error
mae = round(mean_absolute_error(y_test, pred_2),2)
mae
```

```python
#calculating the residual sum of squares in 2d.p
import numpy as np
rss = round(np.sum(np.square(y_test - pred_2)),2)
rss
```

```python
#Root Mean Squared error in 3d.p
from sklearn.metrics import mean_squared_error
rmse = round(np.sqrt(mean_squared_error(y_test, pred_2)), 3)
rmse
```

```python
#Coefficient of Determination in 2d.p
from sklearn.metrics import r2_score
r_2 = round(r2_score(y_test, pred_2),2)
r_2
```

```python
#comparing the effects of regularisation

def get_weights_df(model, feat, col_name):
    
  #this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)  
    return weights_df
```

```python
#looking for lowest weight and highest weight
weight = get_weights_df(model_2, x_train, 'Linear_Model_Weight' )
weight
```

```python
#training a ridge regression model and getting a new rmse
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_train, y_train)
ridge_pred = ridge_reg.predict(x_test)
rmse_2 = round(np.sqrt(mean_squared_error(y_test, ridge_pred)), 3)
rmse_2
```

```python
#training a lasso regression model
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train,y_train)
lasso_pred = lasso_reg.predict(x_test)
```

```python
#obtaining new feature weights and checking for non_zero features
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
lasso_weights_df
```

4 feature have non_zero weights

```python
#rmse with the lasso regression
rmse_3 =  round(np.sqrt(mean_squared_error(y_test, lasso_pred)), 3)
rmse_3
```
